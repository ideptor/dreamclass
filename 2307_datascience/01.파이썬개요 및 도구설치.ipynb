# 파이썬 역사

## 1991년 탄생(0.9.0)
`파이썬(Python)`은 `귀도 반 로섬(Guido van Rossum)`이 1989년 크리스마스 휴가 기간에 취미로 시작하여  
`1991년`에 `최초 버전인 0.9.0`을 발표한 프로그래밍 언어입니다.  
파이썬의 이름은 코미디 프로그램인 "Monty Python's Flying Circus"에서 영감을 받아 지어졌습니다.
https://ko.wikipedia.org/wiki/%ED%8C%8C%EC%9D%B4%EC%8D%AC

## 1994년 (1.0)
1994년 1.0 버전이 출시되며, 파이썬은 점차 커뮤니티와 사용자들에게 인기를 얻어가기 시작했습니다.   
파이썬은 `C 언어로 작성된 인터프리터 언어`로서, 간결하고 읽기 쉬운 문법이 특징입니다.

## 2000년 (2.0)
2000년에 파이썬 2.0 버전이 출시되면서 유니코드 지원 등 새로운 기능들이 추가되었습니다.  
2.x 버전 시대 동안에는 파이썬이 점차 커뮤니티에서 널리 사용되기 시작했습니다.

## 2008년 (3.0)
파이썬 3.0 버전은 파이썬 언어의 개선된 버전으로, 언어의 불필요한 요소들을 정리하고 더욱 향상된 기능들을 제공했습니다.  
하지만 이로 인해 기존에 작성된 `파이썬 2.x 버전의 코드들과 호환성 문제`가 발생하여,  
초기에는 많은 개발자들이 3.x 버전을 채택하지 않았습니다.

## 2010년 (폭발적인 성장)
`데이터 과학`, `인공지능`, `웹 개발` 등 다양한 분야에서 파이썬의 사용이 급증하였습니다.  
파이썬의 라이브러리와 프레임워크들이 발전하여 파이썬이 `데이터 분석` 및 `기계 학습` 분야에서 주류 언어로 자리 잡게 되었습니다.

## 2020년 (파이썬 인기 지속)
파이썬은 여전히 `세계적으로 가장 인기 있는 프로그래밍 언어 중 하나`이며, 개발자들과 기업들에게 많은 사랑을 받고 있습니다.  
2020년대에는 파이썬 3.x 버전을 사용하는 것이 권장되며,   
파이썬 2.x 버전은 2020년 1월에 공식적으로 지원 종료되었습니다.

# 파이썬 프로그래밍 특징

- `간결하고 읽기 쉬운 문법`: 파이썬은 직관적이고 간결한 문법을 가지고 있어 코드를 읽고 이해하기 쉽습니다. 가독성이 좋아서 코드의 유지보수와 협업에 매우 용이합니다.

- `인터프리터 언어`: 파이썬은 C나 C++과 같은 컴파일 언어와 달리 인터프리터 언어입니다. 따라서 개발자가 코드를 작성하고 바로 실행해볼 수 있어 빠른 개발과 테스트가 가능합니다.

- `다양한 운영체제 지원`: 파이썬은 여러 플랫폼에서 실행될 수 있으며, Windows, macOS, Linux 등 다양한 운영체제에서 동작합니다.

- `동적 타이핑(Dynamic Typing)`: 파이썬은 변수의 데이터 타입을 미리 선언하지 않고, 변수가 할당되는 값에 따라 자동으로 데이터 타입이 결정됩니다. 이로 인해 빠르게 개발할 수 있습니다.

- `객체 지향 프로그래밍 지원`: 파이썬은 객체 지향 프로그래밍을 지원하여 코드의 재사용성과 모듈화를 높여줍니다.

- `풍부한 라이브러리와 모듈`: 파이썬은 많은 표준 라이브러리와 서드파티 라이브러리가 존재하여 다양한 작업을 더 쉽게 수행할 수 있습니다. 데이터 분석, 웹 개발, 인공지능, 자연어 처리 등 다양한 분야를 지원합니다.

- `커뮤니티와 생태계`: 파이썬은 활발한 개발자 커뮤니티와 풍부한 오픈 소스 프로젝트로 구성된 생태계를 가지고 있습니다. 이를 통해 개발자들은 문제를 해결하는데 도움을 받고, 다른 사람들과 협력하여 프로젝트를 진행할 수 있습니다.

- `쉬운 배우기`: 파이썬은 처음 프로그래밍을 배우는 사람들에게도 접근하기 쉬운 언어입니다. 진입 장벽이 낮아 프로그래밍 초보자들도 쉽게 시작할 수 있습니다.

## 1. 가상환경 설정

### 가상환경이란

파이썬 가상환경(Virtual Environment)은 프로젝트마다 독립적인 환경을 구축하여  
각 프로젝트가 필요로 하는 패키지들을 격리된 상태에서 관리할 수 있도록 도와주는 도구입니다.  
이를 통해 프로젝트 간의 패키지 의존성 충돌을 피하고, 각각의 프로젝트가 자신만의 환경을 유지할 수 있습니다.

파이썬 가상환경을 사용하는 이유:

1. 의존성 충돌 회피: 서로 다른 프로젝트에서 사용하는 패키지들이 충돌하지 않도록 격리된 환경을 제공합니다.

2. 프로젝트 별 패키지 버전 관리: 프로젝트마다 필요로 하는 패키지의 버전을 독립적으로 관리하여, 각 프로젝트가 안정적으로 동작하도록 합니다.

3. 가벼운 가상화: 파이썬 가상환경은 가볍고 빠르게 생성되며, 필요한 패키지만 설치하여 자원을 절약합니다.


### 가상환경을 만들고 관리하는 방법:

--- 

**1-1. 가상환경 생성(venv 모듈 이용)**

파이썬 3.3 이상의 버전에서는 venv 모듈을 사용하여 가상환경을 생성할 수 있습니다.  
가상환경을 생성하려는 프로젝트 폴더로 이동한 후, 다음 명령을 실행합니다:

```bash
python3 -m venv {가상환경이름}
```

---

**1-2. 가상환경 활성화**

가상환경을 생성하면 해당 가상환경을 활성화해야 합니다.  
이렇게 하면 프로젝트 폴더에서 가상환경에 설치한 패키지들을 사용할 수 있습니다.  
가상환경을 활성화하는 방법은 운영체제에 따라 다릅니다:

#### Windows:

```bash
{가상환경이름}\Scripts\activate
```

#### macOS 또는 Linux:

```bash
source {가상환경이름}/bin/activate
```


가상환경이 활성화 되면 prompt에 가상환경이름이 보여짐

```
(myvenv) C:\Users\school>
```


---


**1-3. 가상환경 비활성화**

가상환경을 사용한 작업을 마치면 환경을 비활성화하여 원래의 파이썬 환경으로 돌아갈 수 있습니다.  
가상환경에서 비활성화하는 방법은 다음과 같습니다:

```bash
(myvenv) C:\Users\school>deactivate
C:\Users\school>
```
