## 7.1 값을 여러개 출력하기

`print` 에는 변수나 값 여러개를 `,`(콤마) 로 구분하여 넣을 수 있습니다.

```python
print(1,2,3)
print("Hello", "Python")
```

````python
print(1,2,3, sep="*****")
print("Hello", "Python", sep="      ")
````

````python
print(1,2,3)
print("Hello", "Python")
````

````python
print(1,2,3)
````

````python
print("Hello", "Python")
````

### 7.1.1 sep= 로 값 사이에 문자 넣기
기본은 " " 을 넣도록 되어 있습니다.  
`sep` 은 `separate` 의 의미입니다.

```python
print(1,2,3, sep=",")
print("Hello", "World", sep="")
print(1920,1080,sep="x")
```

````python
print(1,2,3, sep=",")
````

````python
print("Hello", "World", sep="")
````

````python
print(1920,1080,sep=" x ")
````

````python
print("ab")
````

````python
print("abcd\b\bb")
````

````python
print("a'b") 
print('a"b') 
````

````python
print("a\"b")  #>  "
print("a\'b")  #>  ' 
print("a\nb")  #>  new line
print("a\tb")  #>  tab (4칸띄우기)
print("a\\b")  #>  \
````

````python
print("hi\"hello") 
print("hello\thello") 
````

## 7.2 줄바꿈 활용하기
줄바꿈문자 `\n` 를 사용합니다.
```python

```

### 7.2.1 end= 사용하기

참고: `print()`는 기본적으로 출력하는 값 끝에 `\n`을 붙입니다.  
  
① 어떤 출력이 나올까요?
```python
print(1)
print(2)
print(3)
```

② 어떤 출력이 나올까요?
```python
print(1, end="")
print(2, end="")
print(3)
```

③ 어떤 출력이 나올까요?
```python
print(1, end=" ")
print(2, end=" ")
print(3)
```

````python
print(1)
print(2)
print(3)
````

````python
print(1,2,3, end="\n", sep="(((")
print(4, end="")
print(5)
````

````python
print(1, end=" (^.^))))~ ")
print(2, end=" !@#%$^& ")
print(3)
````

## 7.5 심사문제: 날짜와 시간 출력하기

````python
year, month, day, hour, minute, second = input().split()

print(year,month,day,sep="-",end="T")
print(hour, minute, second, sep=':')
````

````python
year, month, day, hour, minute, second = input().split()
print(year,"-",month,"-",day,"T",hour,":",minute,":",second, sep="")
````

````python
year, month, day, hour, minute, second = input().split()

print(year, month, day, sep='-', end="T")
print(hour, minute, second, sep=':')
````
